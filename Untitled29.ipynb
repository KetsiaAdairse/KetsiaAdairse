```python
import random
import pandas as pd

def simular_fissao(iteracoes=50,
                   neutrons_iniciais=5,
                   prob_fissao=0.7,
                   neutrons_por_fissao=2,
                   energia_por_fissao=200):

    neutrons = neutrons_iniciais
    dados = []

    for i in range(iteracoes):
        fissões = 0

        for _ in range(neutrons):
            if random.random() < prob_fissao:
                fissões += 1

        energia = fissões * energia_por_fissao
        novos_neutrons = fissões * neutrons_por_fissao

        dados.append({
            "iteracao": i,
            "neutrons": neutrons,
            "fissoes": fissões,
            "energia": energia
        })

        neutrons = novos_neutrons

    return pd.DataFrame(dados)

df = simular_fissao()
df.to_csv("simulacao_fissao.csv", index=False)
```

```python
import random
import pandas as pd

def simular_fissao(
    iteracoes=50,
    neutrons_iniciais=5,
    prob_fissao=0.7,
    neutrons_por_fissao=2,
    energia_por_fissao=200,
    fator_controle=0.3  # NOVO
):

    neutrons = neutrons_iniciais
    resultados = []

    for i in range(iteracoes):
        fissoes = 0

        for _ in range(neutrons):
            if random.random() < prob_fissao:
                fissoes += 1

        energia = fissoes * energia_por_fissao

        novos_neutrons = fissoes * neutrons_por_fissao

        # 🔽 AQUI ESTÁ O CONTROLE
        neutrons_controlados = int(novos_neutrons * (1 - fator_controle))

        resultados.append({
            "iteracao": i,
            "neutrons": neutrons,
            "fissoes": fissoes,
            "energia": energia,
            "controle": fator_controle
        })

        neutrons = neutrons_controlados

        # evita loop infinito gigante
        if neutrons > 1_000_000:
            break

    return pd.DataFrame(resultados)
```

```python
cenarios = [
    {"nome": "Subcritico", "controle": 0.8},
    {"nome": "Critico", "controle": 0.5},
    {"nome": "Supercritico", "controle": 0.1},
]

df_final = pd.DataFrame()

for c in cenarios:
    df_temp = simular_fissao(fator_controle=c["controle"])
    df_temp["cenario"] = c["nome"]
    df_final = pd.concat([df_final, df_temp])

df_final.to_csv("simulacao_completa.csv", index=False)
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# estilo mais bonito
sns.set(style="whitegrid")

# 🔥 1. Energia ao longo do tempo
plt.figure()
sns.lineplot(data=df_final, x="iteracao", y="energia", hue="cenario")
plt.title("Energia Gerada por Cenário")
plt.xlabel("Iteração")
plt.ylabel("Energia")
plt.show()

# 🔥 2. Crescimento de nêutrons
plt.figure()
sns.lineplot(data=df_final, x="iteracao", y="neutrons", hue="cenario")
plt.title("Crescimento de Nêutrons")
plt.xlabel("Iteração")
plt.ylabel("Nêutrons")
plt.show()

# 🔥 3. Total de energia por cenário
plt.figure()
sns.barplot(data=df_final, x="cenario", y="energia")
plt.title("Comparação de Energia Total")
plt.show()
df_total = df_final.groupby("cenario")["energia"].sum().reset_index()
```
