```python
import pandas as pd

url = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science/refs/heads/main/base-de-dados-challenge-1/loja_1.csv"
url2 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science/refs/heads/main/base-de-dados-challenge-1/loja_2.csv"
url3 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science/refs/heads/main/base-de-dados-challenge-1/loja_3.csv"
url4 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science/refs/heads/main/base-de-dados-challenge-1/loja_4.csv"

loja = pd.read_csv(url)
loja2 = pd.read_csv(url2)
loja3 = pd.read_csv(url3)
loja4 = pd.read_csv(url4)

loja.head()
```

```python
# Adicionando uma coluna para identificar a origem de cada dado
loja['Loja'] = 'Loja 1'
loja2['Loja'] = 'Loja 2'
loja3['Loja'] = 'Loja 3'
loja4['Loja'] = 'Loja 4'

# Concatenando todas as tabelas em uma só
df_vendas = pd.concat([loja, loja2, loja3, loja4], ignore_index=True)

# Ajustando o formato da data para garantir que os cálculos temporais funcionem
df_vendas['Data da Compra'] = pd.to_datetime(df_vendas['Data da Compra'], dayfirst=True)

print("Tabelas unificadas com sucesso!")
```

```python
# Analise de Faturamento

faturamento_total = df_vendas['Preço'].sum()
faturamento_por_loja = df_vendas.groupby('Loja')['Preço'].sum().sort_values(ascending=False).reset_index()

print(f"Faturamento Total: R$ {faturamento_total:,.2f}")
print("\nFaturamento por Unidade:")
display(faturamento_por_loja)
```

```python
# Vendas por Categoria

vendas_categoria = df_vendas.groupby('Categoria do Produto')['Preço'].sum().sort_values(ascending=False).reset_index()
print("Faturamento por Categoria:")
display(vendas_categoria)
```

```python
# Média de Avaliação das Lojas

avaliacao_media = df_vendas.groupby('Loja')['Avaliação da compra'].mean().sort_values(ascending=False).reset_index()
print("Média de Avaliação por Loja (escala 1-5):")
display(avaliacao_media)
```

```python
# Produtos mais e menos Vendidos

mais_vendidos = df_vendas['Produto'].value_counts().head(5)
menos_vendidos = df_vendas['Produto'].value_counts().tail(5)

print("Top 5 Produtos Mais Vendidos:")
print(mais_vendidos)
print("\n5 Produtos Menos Vendidos:")
print(menos_vendidos)
```
